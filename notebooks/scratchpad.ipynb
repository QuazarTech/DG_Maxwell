```python
def print1D(array):
    '''
    Variables:
        array = 1D numpy array
    '''
    if len(array.shape) != 1:
        print('Cannot print the array.')
        return
    
    print('[', end = '')
    for element_idx, element in enumerate(array):
        if element_idx == (array.shape[0] - 1):
            print(element, end = '],\n')
            pass
        else:
            print(element, end = ',')
            pass
        pass
    return
```

```python
import numpy as np
from scipy import special as sp

def LGL_points(N):
    """
    Takes N as an interger input and returns N LGL points.
    """
    poly_x = np.poly1d([1, 0])
    temp = (N) * (poly_x * sp.legendre(N - 1) - sp.legendre(N))
    lgl = temp.r
    lgl = np.sort(lgl)
    return lgl


N = 20
for n in np.arange(2, N):
    xi_LGL = LGL_points(int(n))
    print(n)
    print1D(xi_LGL)
    pass
```
